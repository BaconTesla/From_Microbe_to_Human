```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
x_1 = np.load('./datas/FengQ_control_vs_CRC.npy')

with open('./datas/info.txt', 'r') as f:
    y = np.array([line.strip() for line in f])    

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)


x_train, x_test, y_train, y_test = train_test_split(x_1, y_encoded, test_size=0.3)

clf = RandomForestClassifier()

param = {"n_estimators": [100,200,300,500,800,1200], "max_depth": [5,8,15,25,30]}

gc = GridSearchCV(clf, param_grid=param, cv=2)

gc.fit(x_train, y_train)

print("随机森林预测的准确率为：", gc.score(x_test, y_test))
print("最佳的模型为：", gc.best_estimator_)
print("交叉验证的结果：", gc.cv_results_)
```

```python
X1 = np.load('./datas/FengQ_control_vs_CRC.npy')
X2 = relative_abundance.to_numpy()

with open('./datas/info.txt', 'r') as f:
    y = np.array([line.strip() for line in f])  

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# 对数据集1
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y_encoded, test_size=0.3, random_state=42)

# 对数据集2
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y_encoded, test_size=0.3, random_state=42)
```

```python
# 模型训练

# 数据集1的模型
rf1 = RandomForestClassifier(random_state=42)
rf1.fit(X1_train, y1_train)

# 数据集2的模型
rf2 = RandomForestClassifier(random_state=42)
rf2.fit(X2_train, y2_train)
```

```python
# 模型评估
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 数据集1的预测和评估
y1_pred = rf1.predict(X1_test)
accuracy1 = accuracy_score(y1_test, y1_pred)
precision1 = precision_score(y1_test, y1_pred, average='weighted')
recall1 = recall_score(y1_test, y1_pred, average='weighted')
f1_1 = f1_score(y1_test, y1_pred, average='weighted')

# 数据集2的预测和评估
y2_pred = rf2.predict(X2_test)
accuracy2 = accuracy_score(y2_test, y2_pred)
precision2 = precision_score(y2_test, y2_pred, average='weighted')
recall2 = recall_score(y2_test, y2_pred, average='weighted')
f1_2 = f1_score(y2_test, y2_pred, average='weighted')
```

```python
# 交叉验证
from sklearn.model_selection import cross_val_score

# 数据集1的交叉验证
scores1 = cross_val_score(rf1, X1, y_encoded, cv=5, scoring='accuracy')

# 数据集2的交叉验证
scores2 = cross_val_score(rf2, X2, y_encoded, cv=5, scoring='accuracy')
```

```python
# 超参数调优
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# 数据集1的超参数调优
grid_search1 = GridSearchCV(rf1, param_grid, cv=5, scoring='accuracy')
grid_search1.fit(X1_train, y1_train)
best_rf1 = grid_search1.best_estimator_

# 数据集2的超参数调优
grid_search2 = GridSearchCV(rf2, param_grid, cv=5, scoring='accuracy')
grid_search2.fit(X2_train, y2_train)
best_rf2 = grid_search2.best_estimator_
```

```python
# 再次评估模型
# 数据集1的最佳模型评估
y1_best_pred = best_rf1.predict(X1_test)
best_accuracy1 = accuracy_score(y1_test, y1_best_pred)
# 可继续计算其他指标

# 数据集2的最佳模型评估
y2_best_pred = best_rf2.predict(X2_test)
best_accuracy2 = accuracy_score(y2_test, y2_best_pred)
```

```python
# 结果比较
print("数据集1的准确率:", best_accuracy1)
print("数据集2的准确率:", best_accuracy2)
```

```python
# 假设检验，判断模型性能差异的显著性
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(scores1, scores2)
print("t-statistic:", t_stat)
print("p-value:", p_value)
```
